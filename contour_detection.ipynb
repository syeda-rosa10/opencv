```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```
img=cv2.imread('/content/drive/My Drive/opencv/download.jpg')
```

```
print(img.shape)
cv2_imshow(img)
```

```
gray1=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
print(gray1.shape)
cv2_imshow(gray1)
```

```
_,thres1=cv2.threshold(gray1,85,255,cv2.THRESH_BINARY)
print(thres1.shape)
cv2_imshow(thres1)
```

```
contours, hierarchy = cv2.findContours(thres1, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
print("Number of Contours is: " + str(len(contours)))
```

```
image = cv2.drawContours(img, contours, -1, (0, 255, 0), 2)
```

```
cv2_imshow(image)
```
